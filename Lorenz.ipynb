```python
import numpy as np
import matplotlib.pyplot as plt

import plotly.express as px

from lorenz_system import LorenzSystem
from rossler_system import RosslerSystem
from duffing_system import DuffingSystem
```

```python
# system = LorenzSystem(sigma=10, r=28, b=8/3)
system = LorenzSystem(sigma=10, r=15, b=8/3)

time = 50
step = 0.01
n_steps = int(time / step)

x, time = system.solve(
    # xyz_0 = np.array([10., 10., 10.]),
    # xyz_0 = np.array([8.5, 7.8, 23]),
#     xyz_0 = [8.5, 7.8, 100],
    xyz_0 = [8.5, 7.8, 23.],
    step=step,
    n_steps=n_steps
)
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

ax.plot(x[:,0], x[:,1], x[:,2])
ax.plot(x[:,0], x[:,1], zs=0, zdir='z')
```

```python
# plt.plot(x[:, 0], x[:, 1])
px.line(x=x[:, 0], y=x[:, 1])
```
